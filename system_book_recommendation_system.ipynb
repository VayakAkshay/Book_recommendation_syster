```python
popular_df = popular_df[popular_df["nums_rating"]>260].sort_values("avg_rating",ascending=False).head(50)
```

```python
popular_books = popular_df.merge(books_data,on="Book-Title").drop_duplicates("Book-Title")
```

```python
x = ratings_data.groupby("User-ID").count()["Book-Rating"] > 200
best_user = x[x].index
best_user
```

```python
filter_rating = ratings_data[ratings_data["User-ID"].isin(best_user)]
```

```python
filter_rating
```

```python
y = filter_rating.groupby("Book-Title").count()["Book-Rating"] >= 50
best_books = y[y].index
best_books
```

```python
final_rating = filter_rating[filter_rating["Book-Title"].isin(best_books)]
```

```python
final_rating
```

```python
pt = final_rating.pivot_table(index = "Book-Title",columns='User-ID',values = 'Book-Rating')
```

```python
pt.isnull().sum() 
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(pt)
```

```python
def recommend(book):
    index = np.where(pt.index==book)[0][0]
    similar_items = sorted(list(enumerate(similarity[index])),key = lambda x:x[1],reverse = True)[1:6]
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend("1st to Die: A Novel")
```
